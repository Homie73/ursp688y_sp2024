```python
import folium
import csv
import os

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Set the working directory
os.chdir('/content/drive/Shareddrives/URSP 688Y/Exercice7')

# File path to your CSV
filename = '/content/drive/Shareddrives/URSP 688Y/Exercice7/metro_stops_montgomery_county.csv'

# Path to the GeoJSON file
geojson_file = 'Montgomery_County_Boundary.geojson'

# Create a map centered around Montgomery County with interactive controls so that we can zoom in and see more clearly
montgomery_map = folium.Map(
    location=[39.0, -77.1],
    zoom_start=12.5,
    tiles='OpenStreetMap',
    zoom_control=True,
    scrollWheelZoom=False,
    dragging=True
)

# Add Montgomery County GeoJSON to the map
folium.GeoJson(geojson_file, name='Montgomery County Boundary').add_to(montgomery_map)

# Read data from the CSV file
with open(filename, mode='r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header row
    for station, latitude, longitude in reader:
        latitude, longitude = float(latitude), float(longitude)
        # Create a custom icon (train icon similar to metro icon)
        icon = folium.Icon(color='red', icon='train', prefix='fa')
        # Add a marker with the custom icon for each station
        folium.Marker([latitude, longitude], popup=f"{station}", icon=icon).add_to(montgomery_map)

        # Draw a circle around the station to represent a 15-minute walk area (about 1000 meters)
        folium.Circle(
            location=[latitude, longitude],
            radius=1000,  # radius in meters
            color='red',  # color of the circle's border
            fill=True,     # fill the circle with color
            fill_color='red',  # fill color
            fill_opacity=0.1  # transparency of the fill color
        ).add_to(montgomery_map)

# Display the map
montgomery_map
```

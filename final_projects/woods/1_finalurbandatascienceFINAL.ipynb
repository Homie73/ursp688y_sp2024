```python
# Get % of students enrolled in private school out of total

# Set year as index
percent_df = merged_df.set_index('Year')

# Add column for percentage
percent_df['Percent Enrolled in Private School'] = 0.0

# Iterate over rows and calculate percentage for each year
for Year, row in percent_df.iterrows():
    percent_df.at[Year, 'Percent Enrolled in Private School'] = (row['Private School Enrollment'] / row['Total Enrollment']) * 100

percent_df
```

```python
# Remove percent sign from columns
new_df = percent_df.applymap(lambda x: x.replace('%', '') if isinstance(x, str) else x)

new_df
```

```python
# Calculate correlation coefficient
correlation = new_df['Private School Enrollment'].corr(new_df['Black Graduation Rate'])

print(f"Correlation coefficient between Private School Enrollment and Public Schpp; Black Graduation Rate: {correlation}")
```

```python
# See if its statistically significant
from scipy.stats import pearsonr

# Calculate Pearson correlation coefficient and p-value
correlation, p_value = pearsonr(new_df['Private School Enrollment'], new_df['Black Graduation Rate'])

print(f"Pearson correlation coefficient: {correlation}")
print(f"P-value: {p_value}")
```

**Instructions for use:**

This code contains three functions that produce three complementary sets of results. The functions are as follows:


*   **map_tracts(housing_projects, year, raw_tract_data, raw_race_data):** This function returns a map of public housing units (yellow points) overlaid each census tract's percentage of Black Prince Georgians (blue polygons) for a given year. It takes four arguments: 'housing_projects', 'raw_tract_data', and 'raw_race_data' must be files contained in the larger folder, but 'year' can simply be filled in with 1990, 2000, 2010, or 2020.
*   **final_dataframe(housing_projects, year, raw_tract_data, raw_race_data):** This function returns a dataframe of census tracts that contained public housing for a given year. It also highlights the tracts/rows that meet the threhold for 'unequal distribution' -- that is, tracts that have both a higher than average Black population and density of public housing units. It takes the same arguments as the 'map_tracts' function.
*   **distribution_stats(housing_projects, year, raw_tract_data, raw_race_data):** This function returns a string containing public housing distribution statistics for a given year. Running it for all four years gives a sense of how public housing distrubition has changed over the past 30+ years. It takes the same arguments as the above two functions.

See below for examples of how to use these three functions. The code is currently set to return results for 2020, but it can be edited to show results for 1990, 2000, and 2010 as well.

**Data sources:**


*   Public housing data (PG_county_public_housing.csv) pulled from the National Housing Preservation Database, https://nhpd.preservationdatabase.org/data
*   Racial demographic data for years 1990, 2000, 2010, and 2020 ([year]_race_data.csv) pulled from the IPUMS National Historical Geographic Information System, https://www.nhgis.org/
*   Census tract shapefiles for years 1990, 2000, 2010, and 2020 (PG_county_tracts_[year].shp) pulled from the IPUMS National Historical Geographic Information System, https://www.nhgis.org/
